```python
import pandas as pd
import matplotlib.pyplot as plt
x=[10,12]
y=[2,8]
plt.plot(x, y, marker="*", color= "blue", linestyle="-", linewidth= 5, markersize= 15)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
plt.show()
```

```python
x=[10,12]
y=[2,8]
plt.scatter(x, y,color= "blue", marker="x")
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
plt.show()
```

```python
Name= ["Kalpesh","Nitesh"]
age= [25,20]
plt.bar(Name, age,color= "green")
plt.title("Bar Plot")
plt.xlabel("Name")
plt.ylabel("age")
plt.show()
```

```python
data= [2,3,4,7,8]
plt.hist(data, bins= 5,color= "Yellow", edgecolor="black")
plt.title("Histogram Plot")
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.show()
```

```python
sizes= [20,45,38,69]
labels = ['a','b','c','d']
plt.pie(sizes, labels= labels, autopct= "%1.1f%%", startangle=0)
plt.title('Pie Chart')
plt.show()
```
