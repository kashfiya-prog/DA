```python
import numpy as np

def convert_City_cell(cell):
    if cell=='NaN':
        return 'Hello'
    return cell

def convert_Grade_cell(cell):
    if cell=='NaN':
        return 'C'
    elif cell== '85':
        return 'A'
    return cell

def convert_Name_cell(cell):
    if cell=='20':
        return 'Kalpesh'
    return cell

def convert_Age_cell(cell):
    if cell=='NaN':
        return '21'
    elif cell=='A':
        return '21'
    return cell

def convert_Marks_cell(cell):
    if cell=='NaN':
        return '70'
    elif cell=='Delhi':
        return '95'
    return cell


dm1 = pd.read_csv("new_csv.csv", converters= {"Marks": convert_Marks_cell, "Age": convert_Age_cell,"City": convert_City_cell, "Grade": convert_Grade_cell, "Name": convert_Name_cell})

dm1
```
